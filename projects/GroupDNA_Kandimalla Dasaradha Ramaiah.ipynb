```python
from datetime import datetime
from datetime import timedelta

print("RESPONSE PATTERNS")
print("==============================")
#i got a lot of error so i used colab ai to sovle thats give 1 ,2,3 like comments with code
response_gaps = {}
all_active_dates = {}
prev_time = None
prev_sender =None

# 1. Collect response times and active dates
for i in msg:
    if i["sender"] == "system" or not i["date"] or not i["time"]:
        continue

    current_sender = i["sender"]
    date_clean = i["date"].replace(",", "").strip()
    time_clean = i["time"].strip()
    current_time = datetime.strptime(f"{date_clean} {time_clean}", "%d/%m/%y %H:%M")

    if prev_time is not None and prev_sender != current_sender:
        gap = (current_time - prev_time).total_seconds()
        if 0 <= gap <= 43200:
            if current_sender not in response_gaps:
                response_gaps[current_sender] = []
            response_gaps[current_sender].append(gap)

    if current_sender not in all_active_dates:
        all_active_dates[current_sender] = set()
    all_active_dates[current_sender].add(current_time.date())

    prev_time = current_time
    prev_sender = current_sender

all_chat_dates = [datetime.strptime(d.replace(",", "").strip(), "%d/%m/%y").date() for d in days if d.strip()]
start_date = min(all_chat_dates)
end_date = max(all_chat_dates)
total_timeline_days = (end_date - start_date).days + 1


speed_stats = {}
for person in menbers:
    gaps = response_gaps.get(person, [])
    if gaps:
        speed_stats[person] = sum(gaps) / len(gaps)

if speed_stats:
    fastest = min(speed_stats, key=speed_stats.get)
    slowest = max(speed_stats, key=speed_stats.get)

    #2. Helper to convert seconds into a readable string
    def format_speed(seconds):
        minutes = seconds / 60
        if minutes >= 60:
            return f"{minutes/60:.1f} hours"
        return f"{minutes:.1f} minutes"

    print(f"Fastest replier : {fastest} (avg {format_speed(speed_stats[fastest])})")
    print(f"Slowest replier : {slowest} (avg {format_speed(speed_stats[slowest])})")

#3.Calculate Silent Streaks and track exact date ranges
print("\nLONGEST SILENT STREAKS (days with zero messages)")
streak_data = {}

for person in menbers:
    person_active_days = all_active_dates.get(person, set())
    max_streak = 0
    current_streak = 0

    streak_start = None
    max_streak_start = None
    max_streak_end = None

    for day_offset in range(total_timeline_days):
        check_date = start_date + timedelta(days=day_offset)

        if check_date not in person_active_days:
            if current_streak == 0:
                streak_start = check_date
            current_streak += 1

            if current_streak > max_streak:
                max_streak = current_streak
                max_streak_start = streak_start
                max_streak_end = check_date
        else:
            current_streak = 0

    streak_data[person] = (max_streak, max_streak_start, max_streak_end)

#4. Sort members by their longest streak duration
sorted_streaks = sorted(streak_data.items(), key=lambda x: x[1][0], reverse=True)

# 5. final results
for person, (streak_days, s_start, s_end) in sorted_streaks:
    if streak_days > 0 and s_start and s_end:
        date_range_str = f" ({s_start.strftime('%d %b')} to {s_end.strftime('%d %b')})"
        print(f"{person}: {streak_days} days{date_range_str}")
    elif streak_days > 0:
        print(f"{person}: {streak_days} days")
    else:
        print(f"{person}: 0 days (never went silent)")
```
